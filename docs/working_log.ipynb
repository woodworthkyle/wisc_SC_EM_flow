Reading through this thesis: https://dspace.mit.edu/bitstream/handle/1721.1/61254/701906470-MIT.pdf?utm_source=chatgpt.com

The gist of it is that we need to assign surface impedance to the boundaries of a metal so that the simulator doesn't have to solve for the inside of the metal. The idea being that assigning a sheet impedance imposes a boundary condition for the solver.
- The technical derivations involve treating an incident wave problem as an impedance mismatch problem (ie transmission line problem)
- In "Transmission Lines for Digital and Communication Networks" by Matick it is argued that parallel propagation of a wave (along a conductor) yields the same solutions and surface impedance as that of a normally incident wave.
- Additional derivations are included in: https://library.nrao.edu/public/memos/alma/memo245.pdf
- they go into deriving the surface impedance of a thick conductor
$$Z_s = (1+j)\sqrt{\frac{\omega \mu}{2 \sigma}}$$
- they then show the derivation of a finite thickness conductor (Infinitesimally thin)
$$Z_S = \frac{k}{\sigma}\left[ \frac{e^{kt} + \frac{\sigma Z_{\eta} - k}{\sigma Z_{\eta} + k}e^{-kt}}{e^{kt} - \frac{\sigma Z_{\eta} - k}{\sigma Z_{\eta} + k}e^{-kt}} \right]$$
where, 
$$k = \frac{1+j}{\delta}$$
and simplifies to 
$$Z_s = \frac{k}{\sigma}\left[ \frac{e^{kt} + e^{-kt}}{e^{kt} - e^{-kt}} \right] = \frac{k}{\sigma}coth(kt)$$
when $Z_{\eta}$ is the impedance of free space and frequencies of interest are in the gigahertz regime.
- Great, but how does this actually translate to an EM solve like HFSS? We basically can model a conductor with finite thickness as 2 infinitesimally thin sheets in parallel. 
$$Z_x = \frac{1}{2} \left[ (2Z_S - j\omega L_t) \pm  [4Z_s^2 + (j\omega L_t)^2]^{1/2}\right]$$
where 
$$L_t = \frac{Z_{\eta}tan(\beta t)}{\omega}$$

$$\beta = \omega \sqrt{{\mu}_0 {\epsilon}_0} = \frac{\omega}{c} = \frac{2 \pi}{\lambda}$$

and $Z_s$ is the surface impedance calculated above...I think.
